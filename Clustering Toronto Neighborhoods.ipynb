Import libraries to scrape the web

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from urllib.request import urlopen
```

Open the URL with the Wikipedia page

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
html = urlopen(url)
```

Use Beautiful Soup to parse the HTML by using lxml

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

Get the title of the webpage

```python
title = soup.title.text
print(title)
```

Transfer the table into a Panadas dataframe

```python
Web_Page = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(Web_Page.content,'lxml')
table = soup.find_all('table')[0] 
df = pd.read_html(str(table))[0]
df
```

Drop the cells that don't have an assigned Borough and reset the index values.

```python
df1 = df.dropna()
df1 = df1.reset_index(drop=True)
df1
```

Replace the backslashes with commas.

```python
df2 = df1.replace(' /', ',', regex=True)
df2                          
```
